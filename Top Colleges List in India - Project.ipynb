## Data Cleaning

```python
#1 Replacing the -- with 0 for easy comparision
df = df.replace(to_replace="--",
           value="0")
```

```python
#2 Removing Comma for evaluation
df["UG_fee"] = df["UG_fee"].str.replace(",", "")
df["PG_fee"] = df["PG_fee"].str.replace(",","")
```

```python
#3 Changing The DataType
df = df.astype({'UG_fee':'int','PG_fee':'int','Rating': 'float','Academic':'float','Accommodation':'float','Faculty':'float','Infrastructure':'float','Placement':'float','Social_Life':'float'})
```

```python
#4 Creating a common Overall Scores for Fair Evaluation
Overall_Scores = df[['Rating','Academic','Accommodation','Faculty','Infrastructure','Placement','Social_Life']].mean(axis=1).round(3)
df['Overall_Scores']= Overall_Scores
```

```python
#5 Keeping only entries which have complete information (removing the missing entries)
Colleges_List = df[(df['Rating'] > 0) & (df['Academic'] > 0) & (df['Accommodation'] > 0) & (df['Faculty'] > 0) & (df['Infrastructure'] > 0) & (df['Placement'] > 0) & (df['Social_Life'] > 0) & (df['UG_fee'] != 'N/A') & (df['PG_fee'] != 'N/A') & (df['UG_fee'] != 0) & (df['PG_fee'] != 0)]
Colleges_List
```

## 2) Which are the top colleges in each state based on the overall scores?

```python
# group the DataFrame by State and get the college with the highest Overall_Scores for each state
top_colleges = Colleges_List.loc[Colleges_List.groupby('State')['Overall_Scores'].idxmax()][['State', 'College_Name', 'Overall_Scores']]
top_colleges
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
ax.bar(top_colleges['State'], top_colleges['Overall_Scores'])
ax.set_title('Top College in Each State')
plt.xticks(rotation=90)
ax.set_xlabel('State')
ax.set_ylabel('Overall Score')
plt.show()
```

## 3) Most and Least Number of Colleges Count

```python
# group the data by state and count the number of colleges in each state
colleges_by_state = Colleges_List.groupby('State')['College_Name'].count().reset_index()

# sort the data by the number of colleges in descending order
colleges_by_state = colleges_by_state.sort_values(by='College_Name', ascending=False)

# display the state with the most and least number of colleges
print("State with the most number of colleges:", colleges_by_state.iloc[0]['State'])
print("Number of colleges:", colleges_by_state.iloc[0]['College_Name'])
print("State with the least number of colleges:", colleges_by_state.iloc[-1]['State'])
print("Number of colleges:", colleges_by_state.iloc[-1]['College_Name'])
```

```python
state_counts = Colleges_List.groupby('State')['College_Name'].count()

# Sort the states by the number of colleges in descending order
state_counts = state_counts.sort_values(ascending=False)

# Create a bar chart
plt.subplots(figsize=(15, 6))
plt.bar(state_counts.index, state_counts.values)
plt.xticks(rotation=90)
plt.xlabel('State',size =(12))
plt.ylabel('Number of Colleges')
plt.title('Number of Colleges in Each State')
plt.show()
```

## 4) Average UG and PG fees

```python
Avg_ug_fee = Colleges_List['UG_fee'].mean()
Avg_pg_fee = Colleges_List['PG_fee'].mean()

print("The Average UG fee across all colleges is: Rs {:,.2f}".format(Avg_ug_fee))
print("The Average PG fee across all colleges is: Rs {:,.2f}".format(Avg_pg_fee))
```

## 5) Average UG and PG fees for each state

```python
fees_df = Colleges_List[['State', 'UG_fee', 'PG_fee']]

# Calculate the mean UG and PG fees for each state
fees_by_state = fees_df.groupby('State').mean()

# Reset the index to turn state into a column
fees_by_state = fees_by_state.reset_index()

fees_by_state
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

# plot UG fees
ax.bar(fees_by_state.index, fees_by_state['UG_fee'], color='blue', alpha=0.8, width=0.3, label='UG Fee')

# plot PG fees next to UG fees
ax.bar(fees_by_state.index + 0.3, fees_by_state['PG_fee'], color='red', alpha=0.8, width=0.3, label='PG Fee')

# set labels and title
ax.set_title('Average UG and PG Fees in Each State')
ax.set_xlabel('State')
ax.set_ylabel('Fee (in lakhs)')
ax.legend()

# adjust x-axis
ax.set_xlim([-0.5, len(fees_by_state)-0.5])
ax.set_xticks(range(len(fees_by_state)))
ax.set_xticklabels(fees_by_state.index, rotation=90)

plt.show()
```
